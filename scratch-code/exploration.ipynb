#### DATA

Mobility data from online data table #5 (reference: www.equality-of-opportunity.org/index.php/data).

```python
# load first dataset
mobility = np.loadtxt("../data/absolute-relative-mobility-county.csv", delimiter=",", dtype=str, skiprows=1)

# handle empty strings
mobility[np.where(mobility == '')] = NaN
print mobility[0:10,]
print

# load second data set
locations = np.loadtxt("../data/county-center-longlat.csv", delimiter=",", dtype=str, skiprows=1)
print locations[0:10,]
```

Here, I perform some data transformations:

```python
mobility[0,0]
```

```python
mobility_col = []
sm = shape(mobility)[0]
for i in xrange(sm):
    mobility_col.append(mobility[i,0] + " " + mobility[i,1])
    
location_col = []
sl = shape(locations)[0]
for i in xrange(sl):
    location_col.append(locations[i,2] + locations[i,3])

print mobility_col[0:10]
print location_col[0:10]
```

```python
mobility_arr = np.array(mobility_col).reshape(sm,1)
location_arr = np.array(location_col).reshape(sl,1)

mobility2 = np.hstack((mobility_arr, mobility[:,2:]))
print mobility2[0:10,]
print

location2 = np.hstack((location_arr, locations[:,5:]))
print location2[0:10,]
```

```python
# convert to data frames (pandas)
my_mobility = pd.DataFrame(mobility2, columns=["county", "relative", "absolute"])
my_locations = pd.DataFrame(location2, columns=["county", "lat", "long"])

# merge data sets
alldata = my_mobility.merge(right=my_locations, how="left", on="county")
print pd.DataFrame.head(alldata)
print
                
# convert back to numpy array
data = alldata.values
print data[0:10,]

# extract columns
lat = data[:,3].astype(float)
lon = data[:,4].astype(float)
```

```python
# remove rows of data without lat/lon coordinates
lat_nans = np.where(np.isnan(lat))
lon_nans = np.where(np.isnan(lon))

nan_indices = lat_nans # = lon_nans (they're the same!)

newdata = np.delete(data, nan_indices, axis=0)

# extract columns
county = newdata[:,0]
relative = newdata[:,1].astype(float)
absolute = newdata[:,2].astype(float)
lat = newdata[:,3].astype(float)
lon = newdata[:,4].astype(float)
```

```python
# plot
plt.figure(figsize=[16,8])

plt.subplot(211)
pplt.plot(absolute, "bo", alpha=0.3, linewidth=2, color="blue")
plt.title("Absolute Mobility")
plt.grid(True)
plt.ylabel("Mobility")

plt.subplot(212)
pplt.plot(relative, "bo", alpha=0.3, linewidth=2, color="magenta")
plt.title("Relative Mobility")
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=[8,6])
pplt.scatter(lon, lat, alpha=0.3, s=20, color="brown")
plt.title("Map of county centers with mobility data")
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.grid(True)
plt.show()
```
